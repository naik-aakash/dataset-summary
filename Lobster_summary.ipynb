<h1><center>Dataset overview</center></h1>

At the moment LOBSTER data available for 1520 compounds. 

All the computational data (including raw unprocessed data) is available to download from our Zenodo communuity : https://zenodo.org/communities/automated-bonding-analysis/?page=1&size=20

Dataset size:

- Unprocess raw data : ~345 GB
- Lightweight jsons (Could be directly used for material project website) : ~ 716 MB
- Computational data in pydantic schema (as per atomate2 workflow) : ~80 GB

### Parsers / Plotters - Pymatgen

Parsers for all the LOBSTER output files are already implemented in pymatgen namely : 
- COHPCAR.lobster / COBICAR.lobster / COBICAR.lobster
- ICOHPLIST.lobster / ICOOPLIST.lobster / ICOBILIST.lobster
- DOSCAR.lobster / DOSCAR.LSO.lobster
- MadelungEnergies.lobster 
- CHARGE.lobster

Plotters for DOSCAR, COHP(ICOHP)/COBI (ICOBI)/COOP (ICOOP) are also readily available. Pymatgen DOS plotter now also has option to plot DOS in convention followed in chemistry:
https://github.com/materialsproject/pymatgen/blob/d621490d5d6742b4c7e432d3144dec6858236c79/pymatgen/electronic_structure/plotter.py#L131

### Parsers / Plotters - Lobsterpy

Interactive plotter for COHP/ICOHP are being implemented in Lobsterpy package (Useful for website)

## DOS comparison / Visualizer app  

We have implemented methods in pymatgen that can easily compare DOS using Tanimoto index as similarity measure. (Could be useful to demonstrate it on website) :
- Fingerprint: https://github.com/materialsproject/pymatgen/blob/d621490d5d6742b4c7e432d3144dec6858236c79/pymatgen/electronic_structure/dos.py#L1180
- Compare two DOS fingerprints (Tanimoto index): https://github.com/materialsproject/pymatgen/blob/d621490d5d6742b4c7e432d3144dec6858236c79/pymatgen/electronic_structure/dos.py#L1284

- We also have created a dash app that helps in visualizing the DOS benchmark for our dataset and is available on zenodo : https://doi.org/10.5281/zenodo.7795903
